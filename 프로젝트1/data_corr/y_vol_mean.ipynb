```python
import pandas as pd

# Load the data
data = pd.read_csv('./y_vol_concat.csv', index_col= 0)

# Display the first few rows of the dataframe
data.head()
```

```python
import matplotlib.pyplot as plt

# Remove the first column which is just an index and the 'date' column for calculating averages
data_without_date = data.drop(columns= 'date')

# Calculate the mean for each column
mean_volumes = data_without_date.mean()

# Plotting the means
plt.figure(figsize=(14, 7))
mean_volumes.plot(kind='bar')
plt.title('Average Volume by Time Slot')
plt.xlabel('Time Slots')
plt.ylabel('Average Volume')
plt.xticks(rotation=90)  # Rotate the x-axis labels to show them more clearly
plt.tight_layout()  # Adjust the plot to ensure everything fits without overlapping
plt.show()
```
