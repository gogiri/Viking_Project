```python
import pandas as pd

# 데이터 모두 불러옴
x1 = pd.read_csv('../data/X/X1_KODEX(1d).csv', index_col=0)
x2 = pd.read_csv('../data/X/X2_X3_X4_KODEX_PBR_PER_DY.csv', index_col=0)
x5 = pd.read_csv('../data/X/X5_kosdaq150(1d).csv', index_col=0)
x7 = pd.read_csv('../data/X_minmax/SP_500_MINMAX.csv', index_col=0)
x11 = pd.read_csv('../data/X_minmax/WTI_MINMAX.csv', index_col=0)
x11_1 = pd.read_csv('../data/X/X11_WTI.csv', index_col=0)
x12 = pd.read_csv('../data/X/X12_10Y2Y.csv', index_col=0)
x12_1 = pd.read_csv('../data/X/X12_10Y3M.csv', index_col=0)
x13 = pd.read_csv('../data/X_minmax/VIX_MINMAX.csv', index_col=0)
x14 = pd.read_csv('../data/X_minmax/MINMAX_BIT.csv', index_col=0)
```

```python
y = pd.read_csv('../data/y_minmax/y_minmax_0905.csv')
```

```python
col_list = [x1,x2,x5,x7,x11,x12,x12_1,x13,x14]
```

```python
y['date'] = pd.to_datetime(y['date'], format='%Y%m%d')
x2['date'] = pd.to_datetime(x2['date'], format='%Y%m%d')
x7['date'] = pd.to_datetime(x7['날짜'], format='%Y- %m- %d')
x11['date'] = pd.to_datetime(x11['날짜'], format='%Y- %m- %d')
x13['date'] = pd.to_datetime(x13['날짜'], format='%Y-%m-%d')
x14['date'] = pd.to_datetime(x14['Date'], format='%Y-%m-%d')
```

```python
x7.drop(columns='날짜',inplace=True)
x11.drop(columns='날짜',inplace=True)
x13.drop(columns='날짜',inplace=True)
x14.drop(columns='Date',inplace=True)
```

```python
for i in col_list :
    i['date'] = pd.to_datetime(i['date'], format='%Y-%m-%d')
    y = pd.merge(y, i, on='date', how='inner')
```

```python
y
```

```python
# "K" 제거 및 숫자로 변환
y['WTI_Volume'] = y['WTI_Volume'].str.replace('K', '').astype(float)
```

```python
# Convert all columns (except 'date') to float
for col in y.columns:
    if col != 'date':
        y[col] = y[col].astype(float)
```

```python
y.to_csv('../data/DataFrame_rate/0900_1000_rate.csv')
```
