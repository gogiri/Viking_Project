```python
import pandas as pd
#거래량 만들기
```

```python
df1 = pd.read_csv('../data/y/kodex(5min).csv')
```

```python
# 'date' 컬럼에서 '09xx'을 포함하는 데이터만 필터링한다.
df2 = df1[df1['date'].astype(str).str.endswith('0905')]

df3 = df1[df1['date'].astype(str).str.endswith('0910')]
```

```python
# 'date' 컬럼의 데이터를 YYYYMMDD 형태로 변환
df2['date'] = df2['date'].astype(str).str[:8].astype(int)
df3['date'] = df3['date'].astype(str).str[:8].astype(int)
merged_data = pd.merge(df2, df3, on='date', how='inner')
merged_data
```

```python
merged_data['volume'] = merged_data['volume_x'] + merged_data['volume_y']
```

```python
merged_data[['date','volume']].to_csv('../data/y_volume/y_vol_0900_0910.csv')
```

```python
mdf1 = merged_data[['date','volume']]
df4 = df1[df1['date'].astype(str).str.endswith('1000')]
mdf1['date'] = mdf1['date'].astype(str).str[:8].astype(int)
df4['date'] = df4['date'].astype(str).str[:8].astype(int)
merged_data = pd.merge(mdf1, df4, on='date', how='inner')
merged_data['volume'] = merged_data['volume_x'] + merged_data['volume_y']
merged_data[['date','volume']].to_csv('../data/y_volume/y_vol_0900_1000.csv')
```
