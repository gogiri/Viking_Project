```python
df1 = pd.read_csv('../data/y/kodex(5min).csv')

from datetime import datetime, timedelta

# 시작 시간과 종료 시간 설정
start_time = datetime.strptime("0905", "%H%M")
end_time = datetime.strptime("1530", "%H%M")

# 5분 간격으로 시간 리스트 생성
time_list = []
while start_time <= end_time:
    time_list.append(start_time.strftime("%H%M"))
    start_time += timedelta(minutes=5)

time_list
```

```python
def labeling(time_list):
    for time in time_list:
        # 'date' 컬럼에서 특정 시간을 포함하는 데이터만 필터링
        kodex = df1[df1['date'].astype(str).str.endswith(time)]

        # 'date' 컬럼의 데이터를 YYYYMMDD 형태로 변환
        kodex['date'] = kodex['date'].astype(str).str[:8].astype(int)

        # 'y_volume' 계산
        kodex[f'y_volume_{time}'] = kodex['volume']

        # 결과 저장
        kodex[['date', f'y_volume_{time}']].to_csv(f'../data/y_volume/y_vol_{time}.csv')
```
