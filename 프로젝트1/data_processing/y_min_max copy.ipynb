```python
import pandas as pd
#고가 - 저가
```

```python
df1 = pd.read_csv('../data/y/kodex(5min).csv')
```

```python
# Extract only the time portion from the 'date' column
df1['time'] = df1['date'].astype(str).str[-4:]

# Filter the data to include only rows from 0900 to 0910
filtered_all_dates = df1[(df1['time'] == '0905') | (df1['time'] == '0910') | (df1['time'] == '0915') | (df1['time'] == '0920') | (df1['time'] == '0925') | (df1['time'] == '0930') 
                        | (df1['time'] == '0935')| (df1['time'] == '0940')| (df1['time'] == '0945')| (df1['time'] == '0950')| (df1['time'] == '0955')]#| (df1['time'] == '1000')]

# Drop the 'time' column
filtered_all_dates = filtered_all_dates.drop(columns='time')

filtered_all_dates.head(12)

```

```python
# Group by the date (without time) and calculate the max of 'open' and min of 'low' for each group
grouped_data = filtered_all_dates.groupby(filtered_all_dates['date'].astype(str).str[:-4])
result_optimized = grouped_data.agg(max_high=('high', 'max'), min_low=('low', 'min')).reset_index()

# Rename columns for clarity
result_optimized = result_optimized.rename(columns={'date': 'date_without_time'})

result_optimized.head()
```

```python
result_optimized['date'] = result_optimized['date_without_time']
result_optimized['y_minmax_0955'] = result_optimized['max_high'] - result_optimized['min_low']
```

```python
result_optimized[['date','y_minmax_0955']].to_csv('../data/y_minmax/y_minmax_0955.csv')
```
